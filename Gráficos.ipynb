```python
valoresx = [1,2,3,4]
valoresy = [1,4,2,3]

figura = plt.figure(figsize=(20,3))
figura.suptitle('Titulo Geral')

figura.add_subplot(131)
plt.plot(valoresx, valoresy, label= 'Um dado qualquer')
plt.legend()
plt.title('Grfico 1')

figura.add_subplot(132)
plt.scatter(valoresx, valoresy)
plt.title('Grfico 2')

figura.add_subplot(133)
plt.bar(valoresx, valoresy)
plt.title('Grfico 3')

plt.savefig('Grfico.png')

plt.show(figura)
```

```python
import yfinance as yf

pbr = yf.Ticker("BTC-USD")
```

```python
pbr.info
```

```python
hist = pbr.history(period="max")
hist
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.plot(hist['Close'], label='BTC-USD')
plt.xlabel('Data')
plt.ylabel('Cotao de Fechamentos (USD)')
plt.legend()
plt.show()
```

```python
dividendos = hist['Dividends'].resample('Y').sum()
dividendos
```

```python
plt.plot(dividendos.index.year[:-2], dividendos[:-2], label='BTC-USD') 
plt.xlabel('Data')
plt.ylabel('Dividendos Anuais (USD)')
plt.legend()
plt.show()
```

```python
hist.to_csv('hist.csv', index=True)
```
